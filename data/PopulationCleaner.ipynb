```python
#First read the data set
import requests
import pandas as pd
path_dir = r"C:\Users\schul\Documents\uni\Master_Kogni\Data Literacy Project\Project\Data\UNPopulationData_Raw.csv"

df = pd.read_csv(path_dir)
print(df)
```

```python
# Not so pretty yet, get rid of some columns since we only want the countries and the respective population density according to the years
print(df.columns)

# Remove unnecessary columns
df_slim = pd.DataFrame([df.pop(x) for x in ['Location','Time', 'Value']]).T
print(df_slim)
```

```python
# Lastly, reorder so that the years are our columns
df_years = df_slim.pivot_table (index = 'Location', columns = 'Time', values = 'Value')
print(df_years)

df_years.to_csv('population_data.csv', sep = '\t', encoding = 'utf-8')
```
